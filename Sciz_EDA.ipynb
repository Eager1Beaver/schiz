```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
cobre_df = pd.read_csv("/content/cobre_participants.csv")
mcic_df = pd.read_csv("/content/mcic_participants.csv")
# Combine the two datasets for analysis
comb = [cobre_df, mcic_df]
comb_df = pd.concat(comb)
# Inspect data
print(cobre_df.head())
print("\n\n\n", mcic_df.head(), "\n\n\n")
comb_df.head()
```

```python
# Summary statistics for the combined data set
comb_df.describe()
```

```python
plt.figure(figsize=(8,8))
plt.hist(cobre_df["age"], 20, density=True, color= "blue", alpha=0.7)
plt.xlabel("Age")
plt.ylabel("Distribution")
plt.title("Age Distribution in the COBRE Dataset", fontsize=12)
plt.show()
```

```python
# Get mean age by gender for COBRE
cobre_gen_age = cobre_df.groupby("sex").agg({"age":"mean"}).reset_index()
cobre_gen_age.head()
```

```python
# Plot mean age by gender COBRE
plt.figure(figsize=(8, 4))
sns.barplot(x="sex", y="age", data=cobre_gen_age)
plt.xlabel("Gender")
plt.ylabel("Average Age")
plt.title("Average Age by Gender- COBRE Dataset", fontsize=12)
plt.show()
```

```python
# Get mean age by gender for MCICShare
mcic_gen_age = mcic_df.groupby("sex").agg({"age":"mean"}).reset_index()
mcic_gen_age.head()
```

```python
# Get mean age by gender for MCICShare
mcic_gen_age = mcic_df.groupby("sex").agg({"age":"mean"}).reset_index()
mcic_gen_age.head()
```

```python
# Plot mean age by gender COBRE
plt.figure(figsize=(8, 4))
sns.barplot(x="sex", y="age", data=mcic_gen_age, palette="dark:orange")
plt.xlabel("Gender")
plt.ylabel("Average Age")
plt.title("Average Age by Gender- MCICShare Dataset", fontsize=12)
plt.show()
```

```python
# Get mean age by gender for Full Dataset
comb_gen_age = comb_df.groupby("sex").agg({"age":"mean"}).reset_index()
comb_gen_age.head()
```

```python
# Plot mean age by gender COBRE
plt.figure(figsize=(8, 4))
sns.barplot(x="sex", y="age", data=comb_gen_age, palette="dark:blue")
plt.xlabel("Gender")
plt.ylabel("Average Age")
plt.title("Average Age by Gender- Combined Dataset", fontsize=12)
plt.show()
```

```python
# Define age bins and labels
bins = [0, 20, 30, 40, 50, 60, 65, float('inf')]
labels = ["<20", "20-29", "30-39", "40-49", "50-59", "60-65", "65+"]

# Create a new column with age groups
comb_df["age_group"] = pd.cut(comb_df["age"], bins=bins, labels=labels, right=False)

comb_df.head()
```

```python
# Group gender by age groups
# Group by age_group and gender, then count the occurrences
age_group_gender = comb_df.groupby(["age_group", "sex"], observed=False).size().unstack(fill_value=0)
age_group_gender.head()
```

```python
# Plot of the age distribution by gender for the whole dataset
age_group_gender.plot(kind='bar', stacked=False)
plt.title('Distribution of Age Groups for Each Gender')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Group age groups by diagnosis
# Group by age_ group and gender, then count the occurrences
age_group_dx = comb_df.groupby(["age_group", "dx"], observed=False).size().unstack(fill_value=0)
age_group_dx.head()
```

```python
# Plot of diagnosis by age groups
age_group_dx.plot(kind='bar', stacked=False)
plt.title('Distribution of Age Groups by Diagnosis')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Diagnosis')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Create column for binary diagnosis
comb_df["schiz"] = comb_df["dx"].apply(lambda x: "Schizophrenia" if "Schizophrenia" in x else "No Schizophrenia")
comb_df.head()
```

```python
# Plot the age group distribution by binary diagnosis
age_group_dx_bin = comb_df.groupby(["age_group", "schiz"], observed=False).size().unstack(fill_value=0)
age_group_dx_bin.head()
```

```python
# Plot of binary diagnosis by age group
age_group_dx_bin.plot(kind='bar', stacked=False)
plt.title('Distribution of Age Groups by Binary Diagnosis')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Diagnosis')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Group binary diagnosis by gender
schiz_bin_gender = comb_df.groupby(["sex", "schiz"], observed=False).size().unstack(fill_value=0)
schiz_bin_gender.head()
```

```python
# Plot of binary diagnosis by gender
schiz_bin_gender.plot(kind='bar', stacked=False, colormap="Paired")
plt.title('Distribution of Age Groups by Binary Diagnosis')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Diagnosis')
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
